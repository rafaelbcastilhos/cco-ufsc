Exercício 1. Escreva uma função que conta a frequência de ocorrência de cada palavra em um texto (arquivo txt) e armazena tal quantidade em um dicionário, onde a chave é a palavra considerada.

```python
from collections import Counter

dicionario = {}
with open('./lerolero.txt') as f:
    palavras = f.read().replace(",", "").replace(".", "").lower().split()

dicionario = Counter(palavras)
print(dicionario)
```

Exercício 2. Escreva uma função que apaga do dicionário anterior, todas as palavras que sejam ‘stopwords’. Ver https://gist.github.com/alopes/5358189

```python
with open('./stopwords.txt') as s:
    stopwords = s.read().split()

for stopword in stopwords:
    existe = dicionario.get(stopword)
    if existe != None:
        dicionario.pop(stopword)
    
print(dicionario)
```

Exercício 3. Escreva um programa que lê duas notas de vários alunos e armazena tais notas em um dicionário, onde a chave é o nome do aluno. A entrada de dados deve terminar quando for lida uma string vazia como nome. Escreva uma função que retorna a média do aluno, dado seu nome.

```python
def pesquisar_aluno(dicionario):
    media = 0
    nome = str(input("Pesquisar aluno: "))
    valores = dicionario.get(nome)
    if valores:
        for valor in valores:
            media += valor

    print(f"Media do/a {nome} é: {media / len(valores)}")


def adiciona_nota(dicionario):
    while True:
        notas = []
        nome = str(input("Entre com nome do aluno: "))
        if nome == "":
            break
        if dicionario.get(nome):
            print(f"Ja existe o aluno {nome}")
            break

        nota1 = float(input("Entre com nota 1: "))
        nota2 = float(input("Entre com nota 2: "))
        notas.append(nota1)
        notas.append(nota2)

        dicionario[nome] = notas
    return dicionario


def main():
    dicionario = {}
    dicionario = adiciona_nota(dicionario)
    pesquisar_aluno(dicionario)


main()
```

Exercício 4. Uma pista de Kart permite 10 voltas para cada um de 6 corredores. Escreva um programa que leia todos os tempos em segundos e os guarde em um dicionário, onde a chave é o nome do corredor. Ao final diga de quem foi a melhor volta da prova e em que volta; e ainda a classificação final em ordem (1o o campeão). O campeão é o que tem a menor média de tempos.

```python
import sys


def main():
    piloto_menor_tempo = sys.maxsize
    piloto_menor_tempo_nome = ""
    volta = ""
    dicionario = {}
    tempo_final = {}

    piloto1 = {'Rafael': {
        'Voltas': {'Volta 1': 301, 'Volta 2': 290, 'Volta 3': 288, 'Volta 4': 291, 'Volta 5': 292, 'Volta 6': 287,
                   'Volta 7': 285, 'Volta 8': 300, 'Volta 9': 302, 'Volta 10': 309}}}
    piloto2 = {'Mateus': {
        'Voltas': {'Volta 1': 290, 'Volta 2': 212, 'Volta 3': 205, 'Volta 4': 254, 'Volta 5': 248, 'Volta 6': 278,
                   'Volta 7': 219, 'Volta 8': 240, 'Volta 9': 233, 'Volta 10': 252}}}
    piloto3 = {'Jonata': {
        'Voltas': {'Volta 1': 277, 'Volta 2': 197, 'Volta 3': 209, 'Volta 4': 195, 'Volta 5': 206, 'Volta 6': 209,
                   'Volta 7': 270, 'Volta 8': 258, 'Volta 9': 278, 'Volta 10': 297}}}
    piloto4 = {'Joao': {
        'Voltas': {'Volta 1': 190, 'Volta 2': 288, 'Volta 3': 239, 'Volta 4': 242, 'Volta 5': 228, 'Volta 6': 289,
                   'Volta 7': 261, 'Volta 8': 288, 'Volta 9': 295, 'Volta 10': 279}}}
    piloto5 = {'Paulo': {
        'Voltas': {'Volta 1': 300, 'Volta 2': 219, 'Volta 3': 224, 'Volta 4': 266, 'Volta 5': 222, 'Volta 6': 239,
                   'Volta 7': 253, 'Volta 8': 300, 'Volta 9': 252, 'Volta 10': 266}}}
    piloto6 = {'Lucas': {
        'Voltas': {'Volta 1': 223, 'Volta 2': 364, 'Volta 3': 295, 'Volta 4': 245, 'Volta 5': 212, 'Volta 6': 282,
                   'Volta 7': 238, 'Volta 8': 225, 'Volta 9': 266, 'Volta 10': 227}}}
    dicionario.update(piloto1)
    dicionario.update(piloto2)
    dicionario.update(piloto3)
    dicionario.update(piloto4)
    dicionario.update(piloto5)
    dicionario.update(piloto6)

    for chave, valor in dicionario.items():
        if piloto_menor_tempo > min(valor['Voltas'].values()):
            piloto_menor_tempo = min(valor['Voltas'].values())
            piloto_menor_tempo_nome = chave
            volta = min(valor['Voltas'], key=valor['Voltas'].get)

    def media(valores, tamanho):
        return sum(valores) / tamanho

    for chave, valor in dicionario.items():
        tempoMedio = media(valor['Voltas'].values(), len(valor['Voltas'].values()))
        tempo_final.update({chave: tempoMedio})

    classificacao_final = sorted(tempo_final.items(), key=lambda elem: elem[1])

    print(f"Melhor volta da corrida pelo piloto: {piloto_menor_tempo_nome} na {volta} em {piloto_menor_tempo} segundos")
    print("Classificação final:")
    i = 1
    for chave, valor in classificacao_final:
        print(f"{i} Lugar: {chave} com tempo médio de {valor} segundos.")
        i += 1


main()
```

Exercício 5. Escreva um programa para armazenar uma agenda de telefones usando um dicionário. Cada pessoa pode ter um ou mais telefones e a chave do dicionário é o nome completo da pessoa. Seu programa deve ter as seguintes funções: 
* incluir_novo_nome – essa função acrescenta um novo nome na agenda, com
* um ou mais telefones. Ela deve receber como argumentos o nome e os telefones.
* incluir_telefone – essa função acrescenta um telefone em um nome existente na agenda. Caso o nome não exista na agenda, você deve perguntar se a pessoa deseja incluí-lo. Caso a resposta seja afirmativa, use a função anterior para incluir o novo nome.
* excluir_telefone – essa função exclui um telefone de uma pessoa que já está na agenda. Se a pessoa tiver apenas um telefone, ela deve ser excluída da agenda.
* excluir_nome – essa função exclui uma pessoa da agenda.
* consultar_telefone – essa função retorna os telefones de uma pessoa na agenda

```python
def incluir_nome(dicionario):
    nome = input('Nome completo: ').title()
    telefones = []
    continuar = 'S'
    while continuar == 'S':
        telefone = input('Telefone: ')
        telefones.append(telefone)
        continuar = input('Adicionar mais um número? S/N').upper()
    dicionario[nome] = telefones


def incluir_telefone(dicionario):
    nome = input('Nome completo: ').title()
    if nome in dicionario.keys():
        telefone = input('Telefone: ')
        dicionario[nome].append(telefone)
    else:
        incluir_nome(dicionario)


def excluir_telefone(dicionario):
    nome = input('Nome completo: ').title()
    telefone = input('Telefone: ')
    if nome in dicionario.keys():
        if telefone in dicionario[nome]:
            dicionario[nome].remove(telefone)
        else:
            print('Telefone não cadastrado.')
    else:
        print('Pessoa não cadastrada.')


def excluir_nome(dicionario):
    nome = input('Nome completo: ').title()
    if nome in dicionario.keys():
        del dicionario[nome]
    else:
        print('Pessoa não cadastrada.')


def consulta_telefone(dicionario):
    nome = input('Nome completo: ').title()
    if nome in dicionario.keys():
        print(dicionario[nome])
    else:
        print('Pessoa não cadastrada.')


def main():
    agenda = {}
    while True:
        opcao = int(input(
            '1- Incluir nome \n2- Incluir telefone \n3- Excluir telefone \n4- Excluir nome \n5- Consultar telefones \n6- Sair \nOpção: '))
        while 1 > opcao > 6:
            opcao = int(input('Opção inválida.'))
        if opcao == 6:
            break
        elif opcao == 1:
            incluir_nome(agenda)
        elif opcao == 2:
            incluir_telefone(agenda)
        elif opcao == 3:
            excluir_telefone(agenda)
        elif opcao == 4:
            excluir_nome(agenda)
        elif opcao == 5:
            consulta_telefone(agenda)


main()
```

Exercício 6. Criar 10 frozensets com 30 números aleatórios cada, e construir um dicionário que contenha a soma de cada um deles.

```python
from random import randint
dicionario = {}

for _ in range(10):
    lista = []
    soma = 0
    for _ in range(30):
        aleatorio = randint(0, 1000)
        lista.append(aleatorio)
        soma += aleatorio
    frozen = frozenset(lista)
    dicionario[frozen] = soma

for k, v in dicionario.items():
    print(f"{k}: {v}")
```
