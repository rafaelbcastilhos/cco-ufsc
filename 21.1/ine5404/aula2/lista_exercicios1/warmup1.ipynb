4) Adicione mais dois atributos canal_minimo (valor padrão 1) e canal_maximo (valor padrão 99) e modifique a classe Televisao de forma que, se pedirmos para mudar o canal para baixo, além do mínimo, ela vá para o canal máximo. Se mudarmos para cima, além do canal máximo, que volte ao canal mínimo.

```python
class Televisao:
    def __init__(self, tamanho: int, marca: str, canal: int):
        self.__ligada = False
        self.__canal = canal
        self.__tamanho = tamanho
        self.__marca = marca
        self.__canal_minimo = 1
        self.__canal_maximo = 99

    @property
    def tamanho(self):
        return self.__tamanho

    @property
    def marca(self):
        return self.__marca

    @property
    def canal(self):
        return self.__canal

    @property
    def canal_maximo(self):
        return self.__canal_maximo

    @property
    def canal_minimo(self):
        return self.__canal_minimo

    @canal.setter
    def canal(self, novo_canal: int):
        self.__canal = novo_canal

    def muda_canal_para_cima(self):
        if(self.canal == self.canal_maximo):
            self.canal = self.canal_minimo
        else:
            self.canal+=1

    def muda_canal_para_baixo(self):
        if(self.canal == self.canal_minimo):
            self.canal = self.canal_maximo
        else:
            self.canal-=1

tv1 = Televisao(50, "samsung", 99)
tv2 = Televisao(42, "philco", 1)

print(f"tv1 da marca {tv1.marca} tem tamanho {tv1.tamanho} e esta no canal {tv1.canal}")
print(f"tv2 da marca {tv2.marca} tem tamanho {tv2.tamanho} e esta no canal {tv2.canal}")

tv1.muda_canal_para_cima()
tv2.muda_canal_para_baixo()

print(f"tv1 da marca {tv1.marca} tem tamanho {tv1.tamanho} e esta no canal {tv1.canal}")
print(f"tv2 da marca {tv2.marca} tem tamanho {tv2.tamanho} e esta no canal {tv2.canal}")
```

7) Crie classes para representar estados e cidades. Cada estado tem um nome, sigla e cidades. Cada cidade tem nome e população. Escreva um programa de testes que crie três estados com algumas cidades em cada um. Exiba a população de cada estado como a soma da população de suas cidades.

```python
class Cidade:
    def __init__(self, nome: str, populacao: int):
        self.__nome = nome
        self.__populacao = populacao

    @property
    def nome(self):
        return self.__nome

    @property
    def populacao(self):
        return self.__populacao

class Estado:
    def __init__(self, nome: str, sigla: str, cidades: list):
        self.__nome = nome
        self.__sigla = sigla
        self.__cidades = cidades

    @property
    def nome(self):
        return self.__nome

    @property
    def cidades(self):
        return self.__cidades

    def populacao(self):
        populacao = 0
        for cidade in self.cidades:
#             print(cidade.nome)
            populacao += cidade.populacao
        return populacao
            
caxias = Cidade("Caxias do Sul", 500000)
poa = Cidade("Porto Alegre", 1500000)
rs = Estado("Rio Grande do Sul", "RS", (caxias, poa))

fln = Cidade("Florianopolis", 500000)
palhoca = Cidade("Palhoça", 175000)
sc = Estado("Santa Catarina", "SC", (fln, palhoca))

curitiba = Cidade("Curitiba", 2000000)
londrina = Cidade("Londrina", 570000)
pr = Estado("Paraná", "PR", (curitiba, londrina))

estados = (rs, sc, pr)
for estado in estados:
    print(f"{estado.nome} tem {estado.populacao()} habitantes")
```

9) Escreva classes para as seguintes formas: quadrado, retângulo e círculo.

```python
import math

class Quadrado():
    def __init__(self, lado: float):
        self.__lado = lado
    
    def calcula_area(self):
        self.area = self.__lado * self.__lado


class Retangulo():
    def __init__(self, base: float, altura: float):
        self.__base = base
        self.__altura = altura
        
    def calcula_area(self):
        self.area = self.__base * self.__altura


class Circulo():
    def __init__(self, raio: float):
        self.__raio = raio
        
    def calcula_area(self):
        self.area = math.pi * self.__raio**2
```
