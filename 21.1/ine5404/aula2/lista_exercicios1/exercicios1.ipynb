Exercício 2: Biblioteca. Neste exercício foi desenvolvido as classes Livro e Biblioteca, utilizando listas para armazenas autores e livros respectivamente.

```python
class Livro:
    def __init__(self, titulo: str, ano: int, editora: str, volume: int, autores: []):
        self.__titulo = titulo
        self.__ano = ano
        self.__editora = editora
        self.__volume = volume
        self.__autores = autores
        self.__is_available = True

    @property
    def titulo(self):
        return self.__titulo

    @property
    def ano(self):
        return self.__ano

    @property
    def editora(self):
        return self.__editora

    @property
    def volume(self):
        return self.__volume

    @property
    def autores(self):
        return self.__autores

    @property
    def is_available(self):
        return self.__is_available

    @is_available.setter
    def is_available(self, novo_estado):
        self.__is_available = novo_estado


class Biblioteca:
    def __init__(self, nome):
        self.__nome = nome
        self.__livros = []

    @property
    def nome(self):
        return self.__nome

    @property
    def livros(self):
        return self.__livros

    def adicionar(self, livro: Livro):
        self.__livros.append(livro)

    def retirar(self, livro: Livro):
        livro.is_available = False

    def devolver(self, livro: Livro):
        livro.is_available = True

    def pesquisar(self, pesquisa: str):
        encontrados = []
        print("iniciando a pesquisa...")
        for livro in self.livros:
            if ((pesquisa in livro.titulo or pesquisa in livro.autores) and livro.is_available == True):
                print(f"Livro: {livro.titulo}.")
                print(f"Ano: {livro.ano}.")
                print(f"Editora: {livro.editora}.")
                print(f"Volume: {livro.volume}.")
                print(f"Autores:")
                for autor in livro.autores:
                    print(f" -> {autor};")
                encontrados.append(livro)
        if (len(encontrados) == 0):
            print("Não foi encontrado. :'(")
        print("_____________________________")
        return encontrados


def main():
    livro1 = Livro("Clean Code", 2008, "Alta books", 1, ["Robert Martin"])
    livro2 = Livro("Automatize tarefas maçantes com Python", 2015, "novatec books", 1, ["Al Sweigart"])
    livro3 = Livro("Introdução a Algoritmos", 2012, "Campus Elsevier", 3,
                   ["Thomas H. Cormen", "Charles E. Leiserson", "Ronald L. Rivest", "Clifford Stein"])
    livro4 = Livro("Desenvolvimento Ágil", 2002, "Alta books", 1, ["Robert Martin"])
    livro5 = Livro("UML para programadores Java", 2003, "Alta books", 1, ["Robert Martin"])

    biblioteca = Biblioteca("Biblioteca Central")
    biblioteca.adicionar(livro1)
    biblioteca.adicionar(livro2)
    biblioteca.adicionar(livro3)
    biblioteca.adicionar(livro4)
    biblioteca.adicionar(livro5)
    biblioteca.pesquisar("Clean")
    biblioteca.retirar(livro1)

    biblioteca.pesquisar("Clean Code")
    biblioteca.pesquisar("Automatize tarefas maçantes com Python")
    biblioteca.pesquisar("Thomas H. Cormen")
    biblioteca.devolver(livro1)


main()
```

```python
from abc import ABC, abstractmethod
from datetime import date


class Cliente:
    def __init__(self, nome: str, telefone: str):
        self.__nome = nome
        self.__telefone = telefone

    @property
    def nome(self):
        return self.__nome

    @property
    def telefone(self):
        return self.__telefone


class Conta(ABC):
    def __init__(self, clientes: [], saldo=0, limite=0):
        self.__depositos = []
        self.__saques = []
        self.__saldo = saldo
        self.__limite = limite
        self.__clientes = clientes

    @property
    def saldo(self):
        return self.__saldo

    @saldo.setter
    def saldo(self, novo_saldo: float):
        self.__saldo = novo_saldo

    @property
    def depositos(self):
        return self.__depositos

    @property
    def saques(self):
        return self.__saques

    @property
    def limite(self):
        return self.__limite

    @property
    def clientes(self):
        return self.__clientes

    @abstractmethod
    def sacar(self, valor: float):
        pass

    @abstractmethod
    def depositar(self, valor: float):
        pass

    def adicionar_cliente(self, cliente: Cliente):
        self.__clientes.append(cliente)

    def resumo(self):
        print("_______Resumo_______")
        print(f"Saldo: {self.saldo:.2f}")
        print(f"Clientes:")
        for cliente in self.clientes:
            print(cliente.nome)
        print("____________________")


class ContaCorrente(Conta):
    def __init__(self, clientes: []):
        super().__init__(clientes)

    def depositar(self, valor: float):
        if (valor > 0):
            self.saldo += (valor - 0.05)
            self.depositos.append(
                f"Deposito na conta corrente em {date.today()} no valor de {valor}. Saldo atual: {self.saldo}")
        else:
            print("Valor de deposito não pode ser negativo.")

    def sacar(self, valor: float):
        if (valor > 0):
            if (self.saldo - (valor + 0.10) >= 0):
                self.saldo -= (valor + 0.10)
                self.saques.append(
                    f"Saque na conta corrente em {date.today()} no valor de {valor}. Saldo atual: {self.saldo}")
            else:
                print("Saldo insuficiente.")
        else:
            print("Valor de deposito não pode ser negativo.")


class ContaPoupanca(Conta):
    def __init__(self, clientes: []):
        super().__init__(clientes)

    def depositar(self, valor: float):
        if (valor > 0):
            self.saldo += (valor - 0.15)
            self.depositos.append(
                f"Deposito na conta poupanca em {date.today()} no valor de {valor}. Saldo atual: {self.saldo}")

        else:
            print("Valor de deposito não pode ser negativo.")

    def sacar(self, valor: float):
        if (valor > 0):
            if (self.saldo - (valor + 0.25) >= 0):
                self.saldo -= (valor + 0.25)
                self.saques.append(
                    f"Saque na conta poupanca em {date.today()} no valor de {valor}. Saldo atual: {self.saldo}")
            else:
                print("Saldo insuficiente")
        else:
            print("Valor de deposito não pode ser negativo.")


class ContaEspecial(Conta):
    def __init__(self, clientes: [], limite: float):
        super().__init__(limite, clientes)

    def depositar(self, valor: float):
        if (valor > 0):
            self.saldo += valor
            self.depositos.append(
                f"Deposito na conta especial em {date.today()} no valor de {valor}. Saldo atual: {self.saldo}")
        else:
            print("Valor de deposito não pode ser negativo.")

    def sacar(self, valor: float):
        if (valor > 0):
            if (self.saldo + self.limite - valor >= 0):
                self.saldo -= valor
                self.saques.append(
                    f"Saque na conta especial em {date.today()} no valor de {valor}. Saldo atual: {self.saldo}")
            else:
                print("Saldo e limite insuficiente.")
        else:
            print("Valor de deposito não pode ser negativo.")


def main():
    cliente1 = Cliente("Joao", "00000000")
    cliente2 = Cliente("Jose", "11111111")
    cliente3 = Cliente("Antonio", "22222222")
    cc1 = ContaCorrente([cliente1, cliente2])
    cc1.depositar(500)
    cc1.depositar(200)
    print(cc1.depositos)
    print(cc1.saques)
    cc1.resumo()
    cc1.adicionar_cliente(cliente3)
    cc1.sacar(800)
    cc1.resumo()
    print(cc1.depositos)
    print(cc1.saques)

    cliente4 = Cliente("Pedro", "33333333")
    cp1 = ContaPoupanca([cliente4])
    cp1.depositar(600)
    cp1.sacar(150)
    print(cp1.depositos)
    print(cp1.saques)
    cp1.resumo()

    cliente5 = Cliente("Felipe", "44444444")
    ce1 = ContaEspecial(1500, [cliente5])
    ce1.sacar(760)
    ce1.depositar(500)
    print(ce1.depositos)
    print(ce1.saques)
    ce1.resumo()


main()
```

Exercício 6: Baralho de cartas. Neste exercício utilizei a estrutura de dados pilha para representar o baralho. Também implementando o método __str__ para converter os objetos em strings.

```python
import random


class Carta:
    def __init__(self, valor: int, naipe: str):
        self.__valor = valor
        self.__naipe = naipe

    def __str__(self):
        return f'[{self.valor} de {self.naipe}]'

    @property
    def valor(self):
        return self.__valor

    @property
    def naipe(self):
        return self.__naipe


class Baralho:
    def __init__(self):
        self.__cartas = []

        naipes = ["Copa", "Espada", "Ouro", "Pau"]
        for naipe in naipes:
            for i in range(1, 10):
                carta = Carta(i, naipe)
                self.adicionar_carta(carta)

    def __str__(self):
        if len(self.__cartas) == 0:
            return '[VAZIO]'

        string = ''

        for carta in self.__cartas:
            string += str(carta)
            string += ', '

        return string

    def adicionar_carta(self, carta):
        self.__cartas.append(carta)

    def embaralhar(self):
        print("Embaralhando...")
        random.shuffle(self.__cartas)

    def pescar(self):
        return self.__cartas.pop()


def main():
    baralho = Baralho()
    baralho.embaralhar()
    print(baralho)
    print(baralho.pescar())


main()
```

Exercício 8: Reimplemente o jogo da forca completamente orientado a objetos. Neste exercício realizei alguns incrementos e alterações para adaptar para o paradigma orientado a objetos.

```python
import random


class Hangman:
    def __init__(self):
        words = ["ABACAXI", "ABACATE", "LARANJA", "PESSEGO", "MELANCIA", "MAMAO", "CEREJA", "BANANA"]
        random.shuffle(words)
        self.__word = words.pop()
        self.__attempted_letters = []
        self.__incomplete_word = ["_"] * len(self.word)
        self.__mistakes = 0
        self.__has_won = False
        self.__has_lost = False

    @property
    def word(self):
        return self.__word

    @property
    def attempted_letters(self):
        return self.__attempted_letters

    @property
    def incomplete_word(self):
        return self.__incomplete_word

    @property
    def mistakes(self):
        return self.__mistakes

    @property
    def has_won(self):
        return self.__has_won

    @property
    def has_lost(self):
        return self.__has_lost

    def up_mistakes(self):
        self.__mistakes += 1

    def draw(self):
        print("--------------------------------------", end="")
        print(
            "\n    ________                                            " +
            "\n   |        |" +
            "\n   |" + ("        O" if self.mistakes >= 1 else "") +
            "\n   |" + ("        | " if self.mistakes == 2 else "") +
            ("       /| " if self.mistakes == 3 else "") +
            ("       /|\\" if self.mistakes >= 4 else "") + (
                "                        " if self.mistakes < 2 else "              ") +
            "\n   |" + ("       /" if self.mistakes == 5 else "") +
            ("       / \\" if self.mistakes == 6 else "") +
            "\n  / \                "
        )
        print(f"Palavra: ", end='')
        for i in self.incomplete_word:
            print(i + " ", end='')
        print()
        print(f"Tentativas: {self.attempted_letters}")

    def new_attempt(self, attempt: str):
        if not attempt.isalpha():
            print("Isso nem é uma tentativa ¬¬")
            return

        # verifica se a tentativa é uma letra ou palavra
        if len(attempt) == 1:
            # verifica se a tentativa já foi realizada
            if attempt in self.attempted_letters:
                print("Você já tentou essa letra!")
                return

            # adiciona a letra à attempted_letters
            self.attempted_letters.append(attempt)
            self.attempted_letters.sort()

            # busca as ocorrências da letra na palavra final 
            indexes = [i for i, x in enumerate(self.word) if x == attempt]

            # caso não encontre a letra, contabiliza o erro e verifica se o jogo acabou
            if indexes == []:
                self.up_mistakes()
                self.check_lost()
            # tendo encontrado a letra, posiciona-a no(s) seu(s) respectivo(s) lugar(es)
            else:
                for i in indexes:
                    self.incomplete_word[i] = attempt
                self.check_won()

        # se a tentativa é uma palavra, verifica se o jogador ganhou ou errou.
        else:
            if attempt == self.word:
                self.__has_won = True
            else:
                self.up_mistakes()
                self.check_lost()

    def check_won(self):
        if (self.word == ''.join(self.incomplete_word)):
            self.__has_won = True

    def check_lost(self):
        if self.mistakes >= 6:
            self.__has_lost = True


def main():
    hangman = Hangman()
    # laço de repeticao para informar tentativa enquanto não ganhar ou não esgotar todas tentativas
    while True:
        # mostra o as tentativas, palavra a ser descoberta e desenha forca caso tenha erros
        hangman.draw()
        # recebe e verifica a tentativa
        hangman.new_attempt(input('Sua tentativa: ').upper())

        if hangman.has_won:
            print('Parabéns! Você venceu!')
            break

        elif hangman.has_lost:
            # mostra o enforcado com 6 erros
            hangman.draw()
            print('Game over! A palavra era ' + hangman.word)
            break

    print('Foi um prazer jogar com você.')


main()
```

Exercício 9: Campo minado. Projete os atributos e os métodos (sem implementação) necessários para a criação de um campo minado. Neste exercício projetei os atributos e dois métodos simples para realizar uma demonstração de jogo, possibilitando ataque enquanto não explodir nenhuma bomba.

```python
class CampoMinado:
    def __init__(self):
        self.__matriz = [
            [0, 0, 3, 3, 3],
            [2, 0, 0, 4, 0],
            [2, 0, 0, 4, 0],
            [0, 0, 0, 4, 0],
            [2, 2, 0, 4, 0]
        ]
        self.__perdeu = False
    # 0 = limpo
    # 1 = bomba
    # 2 = bomba simples
    # 3 = bomba média
    # 4 = bomba grande
    @property
    def matriz(self):
        return self.__matriz

    @property
    def perdeu(self):
        return self.__perdeu
    
    @perdeu.setter
    def perdeu(self, status):
        self.__perdeu = status

    def ver_campo(self):
        for i in range(len(self.matriz)):
            for j in range(len(self.matriz[i])):
                if (self.matriz[i][j] == 1):
                    print("x", end=" ")
                else:
                    print("~", end=" ")
            print("\n")

    def atacar(self, linha, coluna):
        if self.matriz[linha][coluna] == 0:
            self.matriz[linha][coluna] = 1
            print("Pimba!")
        else:
            print("BOOOOOM!")
            self.perdeu = True



def main():
    cm = CampoMinado()
    perdeu = False
    while not cm.perdeu:
        cm.ver_campo()
        l, c = map(int, input("Entre com coordenadas: ").split())
        cm.atacar(l, c)


main()
```
